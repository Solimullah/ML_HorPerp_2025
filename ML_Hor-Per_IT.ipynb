```python
import numpy as np
import pandas as pd
from scipy.signal import find_peaks

# ------------------------------
# Define Filtering Function
# ------------------------------
def is_proper_signal(sensor_data, peak_height_factor=2.0, min_amplitude=100, std_threshold=20):
    """
    Determines if a row contains a proper signal or just noise.
    
    Parameters:
    - sensor_data: 1D array of sensor values.
    - peak_height_factor: Multiplier for standard deviation to define peak detection threshold.
    - min_amplitude: Minimum required amplitude (max - min) to be considered a valid signal.
    - std_threshold: Minimum required standard deviation to be considered a valid signal.
    
    Returns:
    - True if the row has a valid signal, False if it is noise.
    """
    # Compute amplitude (max - min)
    amplitude = np.max(sensor_data) - np.min(sensor_data)
    
    # Compute standard deviation
    std_dev = np.std(sensor_data)
    
    # Compute dynamic peak threshold
    peak_threshold = np.mean(sensor_data) + peak_height_factor * np.std(sensor_data)
    
    # Detect peaks
    peaks, _ = find_peaks(sensor_data, height=peak_threshold, distance=50)
    
    # Apply filtering conditions:
    return (amplitude > min_amplitude) and (std_dev > std_threshold) and (len(peaks) > 0)

# ------------------------------
# Apply Filtering to Dataset
# ------------------------------
proper_signal_mask = []
for idx in range(len(df)):
    sensor_data = df.iloc[idx, 17:50017].values  # Extract sensor data
    is_valid = is_proper_signal(sensor_data)
    proper_signal_mask.append(is_valid)

# Convert to a boolean Series
proper_signal_mask = pd.Series(proper_signal_mask, index=df.index)

# Keep only rows with proper signals
df_filtered = df[proper_signal_mask].copy()

# ------------------------------
# Summary
# ------------------------------
print("Original number of rows:", len(df))
print("Number of rows with proper signals:", len(df_filtered))
print("Number of noisy rows removed:", len(df) - len(df_filtered))

```

```python
import numpy as np
from scipy.signal import find_peaks

def detect_first_echo(sensor_data, threshold_factor=2.0, min_distance=50):
    """
    Detects the first significant peak (first echo) in the sensor data.
    
    Parameters:
    - sensor_data: 1D numpy array of the signal.
    - threshold_factor: Peak detection threshold (mean + factor * std).
    - min_distance: Minimum distance between peaks.
    
    Returns:
    - First peak index (or None if no peak found).
    """
    threshold = np.mean(sensor_data) + threshold_factor * np.std(sensor_data)
    peaks, _ = find_peaks(sensor_data, height=threshold, distance=min_distance)
    return peaks[0] if len(peaks) > 0 else None  # Return first peak index

# Add 'First_Echo_Index' column if missing
if 'First_Echo_Index' not in df_filtered.columns:
    df_data['First_Echo_Index'] = df_filtered.iloc[:, 17:50017].apply(
        lambda row: detect_first_echo(row.values), axis=1
    )

# Verify it exists now
print(df_data[['First_Echo_Index']].head())
```

# Model testing

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, Flatten, Dense, Dropout, BatchNormalization
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# -------------------------------
# Step 1: Data Preparation
# -------------------------------

# Extract input signals and labels
X = []
y_classification = []
y_distance = []

def extract_first_echo(sensor_data, first_echo_index, window_size=300):
    """Extracts a window around the first echo signal."""
    if pd.isna(first_echo_index):
        return np.zeros(window_size)  # If NaN, return zero signal

    first_echo_index = int(first_echo_index)  # Ensure integer
    start_idx = max(0, first_echo_index - window_size // 2)
    end_idx = min(len(sensor_data), start_idx + window_size)

    extracted_signal = np.zeros(window_size)
    extracted_signal[:end_idx - start_idx] = sensor_data[start_idx:end_idx]
    
    return extracted_signal

# Loop through dataset to extract features
for idx in range(len(df_data)):
    sensor_data = df_data.iloc[idx, 17:50017].values  # Extract sensor data
    first_echo_index = int(df_data.iloc[idx]['First_Echo_Index'])  # Ensure correct column name
    extracted_signal = extract_first_echo(sensor_data, first_echo_index)

    X.append(extracted_signal)
    y_distance.append(df_filtered.iloc[idx]['Distance'])  # Distance regression target
    y_classification.append(df_filtered.iloc[idx][['Object_Box', 'Object_person']].values)  # Object classification

# Convert to numpy arrays
X = np.array(X)
y_distance = np.array(y_distance)
y_classification = np.array(y_classification)  # One-hot encoded labels for classification

# Normalize signals
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Reshape for CNN input
X = X.reshape(X.shape[0], X.shape[1], 1)

# Split dataset into training and testing sets
X_train, X_test, y_class_train, y_class_test, y_dist_train, y_dist_test = train_test_split(
    X, y_classification, y_distance, test_size=0.2, random_state=42
)


print(X_train)
print(X_test)
print(y_class_train)
print(y_class_test)
print(y_dist_train)
print(y_dist_test)
```

```python
# -------------------------------
# Step 2: CNN Model for Classification (Person vs. Object)
# -------------------------------

input_layer = Input(shape=(X.shape[1], 1))

conv1 = Conv1D(filters=64, kernel_size=5, activation="relu")(input_layer)
conv1 = BatchNormalization()(conv1)
pool1 = MaxPooling1D(pool_size=2)(conv1)

conv2 = Conv1D(filters=128, kernel_size=5, activation="relu")(pool1)
conv2 = BatchNormalization()(conv2)
pool2 = MaxPooling1D(pool_size=2)(conv2)

conv3 = Conv1D(filters=256, kernel_size=5, activation="relu")(pool2)
conv3 = BatchNormalization()(conv3)
pool3 = MaxPooling1D(pool_size=2)(conv3)

flatten = Flatten()(pool3)
dense1 = Dense(256, activation="relu")(flatten)
dropout = Dropout(0.4)(dense1)

# Output layer for classification (Person vs Object)
output_class = Dense(y_classification.shape[1], activation="softmax", name="classification_output")(dropout)

# Define classification model
model_classification = Model(inputs=input_layer, outputs=output_class)

# Compile model
model_classification.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

# Print model summary
print("\nModel Summary: Object Classification")
model_classification.summary()

# Train the classification model
history_class = model_classification.fit(
    X_train, y_class_train,
    validation_data=(X_test, y_class_test),
    epochs=30, batch_size=32
)

# Evaluate classification model
test_loss_class, test_acc_class = model_classification.evaluate(X_test, y_class_test)
print(f"\nTest Accuracy (Object Classification): {test_acc_class * 100:.2f}%")

# Save the trained classification model
model_classification.save("cnn_model_classification.h5")

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.signal.windows import hamming  # Fixed Hamming import
from scipy.fftpack import fft
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, classification_report
import joblib  # For saving models

# -------------------------------
# Step 1: Data Preprocessing (ACF, Hamming Window, FFT, PSD)
# -------------------------------

def compute_psd(signal, sampling_rate=512000):
    """ Compute the Power Spectral Density (PSD) of a given signal. """
    if np.all(signal == 0):  # Check if signal is all zeros
        return np.array([]), np.array([])

    window = hamming(len(signal))  
    signal_windowed = signal * window
    
    fft_values = np.abs(fft(signal_windowed))**2
    freqs = np.fft.fftfreq(len(fft_values), 1 / sampling_rate)

    # Extract only frequencies between 30 kHz and 50 kHz
    mask = (freqs >= 30000) & (freqs <= 50000)

    if np.sum(mask) == 0:  # No valid frequencies found
        return np.array([]), np.array([])

    return freqs[mask], fft_values[mask]

# Prepare data storage
X_features = []
y_labels = []

for idx in range(len(df_data)):
    sensor_data = df_data.iloc[idx, 17:50017].values  
    first_echo_index = int(df_data.iloc[idx]['First_Echo_Index'])

    # Extract first echo signal
    window_size = 300
    start_idx = max(0, first_echo_index - window_size // 2)
    end_idx = min(len(sensor_data), start_idx + window_size)
    
    extracted_signal = np.zeros(window_size)
    extracted_signal[:end_idx - start_idx] = sensor_data[start_idx:end_idx]

    # Compute PSD
    freqs, psd_values = compute_psd(extracted_signal)

    if psd_values.size > 0:  # Ensure valid features
        X_features.append(psd_values)
        y_labels.append(1 if df_filtered.iloc[idx]['Object_person'] == 1 else 0)

# Convert to NumPy array only if data is valid
if len(X_features) > 0:
    X_features = np.array(X_features)
    y_labels = np.array(y_labels)
    print("✅ Filtered X_features shape:", X_features.shape)  
else:
    print("⚠ No valid data found after preprocessing. Check dataset or frequency range.")
    exit()  # Stop execution if data is empty

# Standardize features if valid data exists
if X_features.size > 0:
    scaler = StandardScaler()
    X_features = scaler.fit_transform(X_features)
else:
    print("⚠ No valid features found. Skipping StandardScaler.")
    exit()  # Stop execution safely

# -------------------------------
# Step 2: Train MLP Classifier
# -------------------------------

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X_features, y_labels, test_size=0.2, random_state=42)

# Train MLP Classifier
mlp = MLPClassifier(hidden_layer_sizes=(128, 64), activation='relu', solver='adam', max_iter=1000, random_state=42)
mlp.fit(X_train, y_train)

# -------------------------------
# Step 3: Evaluate Model
# -------------------------------

# Predict on test data
y_pred_mlp = mlp.predict(X_test)

# Print Classification Report
print("\nMLP Classification Report:")
print(classification_report(y_test, y_pred_mlp))

# Compute Confusion Matrix
conf_mlp = confusion_matrix(y_test, y_pred_mlp)

# -------------------------------
# Step 4: Fix Confusion Matrix Plotting
# -------------------------------

def plot_confusion_matrix(y_true, y_pred, title="Confusion Matrix"):
    """ Plot a confusion matrix with correct labels. """
    cm = confusion_matrix(y_true, y_pred)
    
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Object', 'Person'], yticklabels=['Object', 'Person'])
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(title)
    plt.show()

# Plot Confusion Matrix
plot_confusion_matrix(y_test, y_pred_mlp, "MLP Confusion Matrix")

# -------------------------------
# Step 5: Save Trained Model for Future Use
# -------------------------------
joblib.dump(mlp, "mlp_person_object_classifier.pkl")
joblib.dump(scaler, "scaler_psd.pkl")  # Save scaler for future normalization
```
