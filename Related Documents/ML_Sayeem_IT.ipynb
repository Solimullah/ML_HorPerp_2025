# Data Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Select the row index to analyze
row_index = 0  # Change this for other rows

# Extract sensor data (D1 to D50000, skipping first 17 columns)
sensor_data = df.iloc[row_index, 17:50017].values  

# Define threshold for echo detection (adjust as needed)
THRESHOLD = np.mean(sensor_data) + 2 * np.std(sensor_data)  # Adaptive threshold

# Find peaks above the threshold
peaks, _ = find_peaks(sensor_data, height=THRESHOLD, distance=50)  # Min distance between peaks

# Get the first detected echo
if len(peaks) > 0:
    first_echo_index = peaks[0]  # First significant peak index
else:
    first_echo_index = None  # No peak detected

# Plot the sensor data
plt.figure(figsize=(12, 6))
plt.plot(sensor_data, label="Sensor Data", linewidth=1.5, color='blue')

# Mark the first echo with a red vertical line
if first_echo_index is not None:
    plt.axvline(x=first_echo_index, color='red', linestyle="--", label=f"First Echo at {first_echo_index}")

# Labels and title
plt.xlabel("Sample Index (Time)")
plt.ylabel("Sensor Value")
plt.title(f"Sensor Data Visualization - Row {row_index}")
plt.legend()

# Show the plot
plt.show()

if first_echo_index is not None:
    print(f"First echo detected at index {first_echo_index}")
else:
    print("No first echo detected. Adjust the threshold or check the data.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Define parameters
num_rows = df.shape[0]  # Total number of rows
THRESHOLD_MULTIPLIER = 2  # Adjust this for sensitivity
first_echo_indices = []  # To store detected first echo positions

# Loop through each row to find the first echo
for row_index in range(num_rows):
    sensor_data = df.iloc[row_index, 17:50017].values  # Extract sensor data (D1 to D50000)

    # Adaptive threshold for each row
    threshold = np.mean(sensor_data) + THRESHOLD_MULTIPLIER * np.std(sensor_data)

    # Find peaks above the threshold
    peaks, _ = find_peaks(sensor_data, height=threshold, distance=50)

    # Store first detected echo index
    if len(peaks) > 0:
        first_echo_indices.append(peaks[0])
    else:
        first_echo_indices.append(None)  # No echo detected for this row

# Convert to numpy array (replace None with NaN for plotting)
first_echo_indices = np.array(first_echo_indices, dtype=np.float32)
first_echo_indices[np.isnan(first_echo_indices)] = np.nan  # Keep NaNs for missing values

# Plot First Echo Positions Across All Rows (Line Graph)
plt.figure(figsize=(12, 6))
plt.plot(range(num_rows), first_echo_indices, color='red', linestyle='-', linewidth=1.5, marker='o', markersize=3, label="First Echo Position")
plt.xlabel("Row Index (Sensor Data)")
plt.ylabel("First Echo Sample Index (Time)")
plt.title("First Echo Detection Across All Rows")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

# Global Spectrogram

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import spectrogram

# Define parameters
num_rows = df.shape[0]  # Total number of rows

# Initialize an array to store averaged spectrogram data
global_sxx = None

# Loop through all rows to compute the spectrogram
for row_index in range(num_rows):
    sensor_data = df.iloc[row_index, 17:50017].values  # Extract sensor data (D1 to D50000)

    # Compute spectrogram for this row
    frequencies, times, Sxx = spectrogram(sensor_data, fs=1000, nperseg=500, noverlap=250)

    # Convert complex values to magnitude
    Sxx_magnitude = np.abs(Sxx)

    # Accumulate the spectrogram data
    if global_sxx is None:
        global_sxx = Sxx_magnitude
    else:
        global_sxx += Sxx_magnitude  # Sum up all spectrograms

# Average the spectrogram across all rows
global_sxx /= num_rows

# Prevent log10(0) by replacing zeros with a small constant
global_sxx[global_sxx == 0] = 1e-10  

# Convert power spectrum to decibels
global_sxx_db = 10 * np.log10(global_sxx)

# Plot the global spectrogram
plt.figure(figsize=(12, 6))
plt.pcolormesh(times, frequencies, global_sxx_db, shading='gouraud', cmap='viridis')

# Labels and title
plt.xlabel("Time (Sample Index)")
plt.ylabel("Frequency (Hz)")
plt.title("Global Spectrogram - Averaged Over All Sensor Data")
plt.colorbar(label="Power/Frequency (dB/Hz)")

# Show the plot
plt.show()
```
